```python
#3 convert list of comments into list of list of tokens & cleaning
def corpus2docs(list_of_docs):
    docs = [doc.replace('\n', ' ').lower() for doc in list_of_docs] # concatenate all paragraphs into 1 line & lower case
    docs = [nltk.word_tokenize(doc) for doc in docs] # tokenize
    docs = [[re.sub(r'\d+', '', w) for w in doc] for doc in docs]# remove digits
    docs = [[w for w in doc if w not in stop_list] for doc in docs] # remove stop-words
    #docs = [[stemmer.stem(w) for w in doc] for doc in docs] # stem the words
    docs = [[w for w in doc if len(w) >= 3] for doc in docs] # retain only words with at least 3 letters
    docs = [[w for w in doc if len(w) <= 15] for doc in docs] # retain words less than 15 letters to remove links 
    return docs
```

```python
#4 convert tokenized comments into word vectors using raw TF
def docs2vecs(docs, dictionary):
    # docs is a list of documents returned by corpus2docs.
    # dictionary is a gensim.corpora.Dictionary object.
    vecs1 = [dictionary.doc2bow(doc) for doc in docs]
    return vecs1
```

```python
def dominant_topic(model, corpus, data):

    topic_df = pd.DataFrame()


    for i, row in enumerate(model[corpus]):
        row = sorted(row, key=lambda x: (x[1]), reverse=True)

        for j, (topic_num, prop_topic) in enumerate(row):
            if j == 0:
                wp = model.show_topic(topic_num)
                topic_keywords = ", ".join([word for word, prop in wp])
                topic_df = topic_df.append(pd.Series([int(topic_num), round(prop_topic,4), topic_keywords]), ignore_index=True)
            else:
                break
    topic_df.columns = ['Dominant_Topic', 'Perc_Contribution', 'Topic_Keywords']

    # Add original text to the end of the output
    contents = pd.Series(data)
    topic_df = pd.concat([topic_df, contents], axis=1)
    return(topic_df)
```
